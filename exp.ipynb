```python
import os
from dotenv import load_dotenv
load_dotenv()
import requests
from openai import OpenAI
EXCHANGERATE_API_KEY = os.getenv("EXCHANGERATE_API_KEY")
GITHUB_TOKEN = os.getenv('GITHUB_TOKEN')
endpoint = "https://models.github.ai/inference"
model_name = "openai/gpt-4o-mini"

client = OpenAI(
    base_url=endpoint,
    api_key=GITHUB_TOKEN,
)

response = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant to convert one currency to another currency.As response create a JSON (dict) with the base curency, amount, target currency and amount in target currency. keep only this JSON (dict) in response",
        },
        {
            "role": "user",
            "content": "convert 250 EUR to USD?",
        }
    ],
    temperature=1.0,
    top_p=1.0,
    max_tokens=1000,
    model=model_name
)

answer = response.choices[0].message.content
# url = f'https://v6.exchangerate-api.com/v6/{EXCHANGERATE_API_KEY}/pair/EUR/GBP/40'

# # Making our request
# response = requests.get(url)
# data = response.json()
# data
```
